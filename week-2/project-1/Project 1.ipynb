```python
# Program to swap ages

name_1 = input('Enter first name:')
age_1 = input('Enter age:')

name_2 = input('Enter second name:')
age_2 = input('Enter age:')

temp = age_1
age_1 = age_2
age_2 = temp

print(f"The age of {name_1} after switching is {age_1}")
print(f"The age of {name_2} after switching is {age_2}")
```
