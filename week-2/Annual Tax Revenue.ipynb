```python
def calculate_atr():
    print("--- Izifin Technology ATR Calculator ---")
    
    # Input block
    try:
        experience = float(input("Enter years of experience: "))
        age = float(input("Enter age of staff: "))
    except ValueError:
        print("Invalid input. Please enter numbers only.")
        return

    # Logic block using if-elif-else
    if experience > 25 and age >= 55:
        atr = 5600000
    elif experience > 20 and age >= 45:
        atr = 4480000
    elif experience > 10 and age >= 35:
        atr = 1500000
    else:
        # This covers staff with < 10 years experience and < 35 years of age
        atr = 550000

    # Output block
    print(f"\nResults for staff:")
    print(f"Experience: {experience} years")
    print(f"Age: {age} years")
    print(f"Annual Tax Revenue (ATR): N{atr:,}")

# Run the program
calculate_atr()
```
