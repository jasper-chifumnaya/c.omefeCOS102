```python
import math

print("Cubic Equation ----- c\nQuartic Equation ----- k\nQuadratic Equation ----- q")

#make sure input is a number
def num(prompt):
    while True:
        x = input(prompt)
        try:
            return float(x)  # If successful, this exits the loop and the function
        except ValueError:
            print("Invalid input. Please enter a numeric value.")


#prompt block for q quadratic equation
def quadratic():
    a = num("Enter value for a:")
    b = num("Enter value for b:")
    c = num("Enter value for c:")
    print(quadcalc(a, b, c))
    
#Quadratic equation calculation
def quadcalc(x, y, z):
    discriminant = pow(y, 2) - 4 * x * z
    if discriminant > 0:
        rooto = (-y + (discriminant**0.5)) / (2 * x)
        roott = (-y - (discriminant**0.5)) / (2 * x)
        ans = (f"There are two real roots are, {rooto} and {roott}")
    elif discriminant == 0:
        root = (-y) / (2 * x)
        ans = (f"There is only one real root, {root}")
    elif discriminant < 0:
        rootr = (-y) / (2 * x)
        rooti = (-discriminant**0.5) / (2 * x)
        ans = (f"There are no real roots, {rootr} and {rooti}")
    return ans


#prompt block for cubic equation
def cubic():
    a = num("Enter value for a:")
    b = num("Enter value for b:")
    c = num("Enter value for c:")
    d = num("Enter value for d:")
    print(cubcalc(a, b, c, d))

#Cubic equation calculation
def cubcalc(x, y, z, w):
    p = (3 * (x * z) - (y**2)) / (3 * (x**2))
    q = ((2 * (y ** 3)) - (9 * x * y * z) + (27 * (x ** 2) * w)) / (27 * (x ** 3))
    discriminant = ((q ** 2) / 4) + ((p ** 3) / 27)
    
    if discriminant > 0:
        u = math.cbrt((-q / 2) + (discriminant**0.5))
        v = math.cbrt((-q / 2) - (discriminant**0.5))
        realroot = u + v - y / 3 * x
        
        compreal = round((-(u + v) / 2) - (y / (3 * x)))
        compimag = round(((3**0.5) / 2) * (u - v))
        #complex roots
        comppos = complex(compreal, compimag)
        compneg = complex(compreal, -compimag)
        ans = (f"There is only one real root, {realroot} and two equal but opposite complex roots, {comppos} and {compneg}")
    
    elif discriminant == 0:
        root1 = (2 * math.cbrt(-q / 2)) - (y / (3 * x))
        root23 = (-math.cbrt(-q / 2)) - (y / (3 * x))
        ans = (f"There are three real roots, {root1}, {root23} and {root23}")

    elif discriminant < 0:
        r = ((-p ** 3) / 27) ** 0.5
        tether = math.acos(-q / (2 * r))
        multi = 2 * ((-p / 3) ** 0.5)
        shift = -y / (3 * x)
        
        root1 = multi * math.cos(tether / 3) - shift
        root2 = multi * math.cos((tether + (2 * math.pi)) / 3) - shift
        root3 = multi * math.cos((tether + (4 * math.pi)) / 3) - shift
        
        ans = (f"There are three roots, {root1}, {root2}, and {root3}")
    
    return ans

    
#prompt block for quartic equation calculation
def quartic():
    a = num("Enter value for a:")
    b = num("Enter value for b:")
    c = num("Enter value for c:")
    d = num("Enter value for d:")
    e = num("Enter value for e:")
    print(quarticcalc(a, b, c, d, e))

def quarticcalc(x, y, z, w, u):
    if x == 0:
        return "This is a cubic equation, choose c instead"
    
    # 1. Normalize coefficients
    A = y / x
    B = z / x
    C = w / x
    D = u / x

    # 2. Depressed quartic: y^4 + py^2 + qy + r = 0
    # Substitution: x = y - A/4
    p = B - (3 * (A**2)) / 8
    q = (A**3 / 8) - (A * B / 2) + C
    r = (-3 * (A**4) / 256) + ((A**2) * B / 16) - (A * C / 4) + D
    
    shift = A / 4

    # 3. Solve the Resolvent Cubic for 'm': 8m^3 + 8pm^2 + (2p^2 - 8r)m - q^2 = 0
    # We call your cubcalc logic specifically for the resolvent coefficients
    # This finds the necessary real root 'm'
    m_val = cubcalc(8, 8 * p, (2 * (p**2) - 8 * r), -(q**2))
    
    # Extract the first numerical value if cubcalc returns a string
    if isinstance(m_val, str):
        # Basic extraction logic for internal calculation
        m = float(m_val.split(",")[1].split("and")[0].strip())
    else:
        m = m_val

    # 4. Ferrari's Quadratic Split
    sqrt_2m = (2 * m)**0.5
    # Denominator grouping is critical here to avoid your previous "5" error
    q_div_2sqrt2m = q / (2 * sqrt_2m)

    # Quadratic 1: y^2 + sqrt(2m)y + (p/2 + m - q/(2*sqrt(2m))) = 0
    q1_b = sqrt_2m
    q1_c = (p / 2) + m - q_div_2sqrt2m
    
    # Quadratic 2: y^2 - sqrt(2m)y + (p/2 + m + q/(2*sqrt(2m))) = 0
    q2_b = -sqrt_2m
    q2_c = (p / 2) + m + q_div_2sqrt2m

    # 5. Final Root Calculation (Using Quadratic Formula)
    def solve_final_quad(qb, qc, s):
        disc = (qb**2) - 4 * qc
        if disc >= 0:
            r1 = (-qb + (disc**0.5)) / 2 - s
            r2 = (-qb - (disc**0.5)) / 2 - s
            return [r1, r2]
        else:
            # Complex roots
            real = -qb / 2 - s
            imag = ((-disc)**0.5) / 2
            return [complex(real, imag), complex(real, -imag)]

    roots = solve_final_quad(q1_b, q1_c, shift) + solve_final_quad(q2_b, q2_c, shift)
    
    ans = f"The four roots are: {roots[0]}, {roots[1]}, {roots[2]}, and {roots[3]}"
    return ans
        
    
choice = (input(" Input key for the equation you'd like to calculate:"))

match choice:
    case "q":
        quadratic()
    case "c":
        cubic()
    case "k":
        quartic()
    case _:
        print("Wrong Input")
```
