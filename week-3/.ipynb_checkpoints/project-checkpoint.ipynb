```python
import tkinter as tk
from tkinter import messagebox as msgbox
import pandas as pd
```

```python
# Load employees data
df = pd.read_csv("GIG-logistics(in).csv")

# checking employees data

def c_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    # filtering
    employee = df[(df['Name'].str.lower() == name.lower()) & (df['Department'].str.lower() == dept.lower())]
    if not employee.empty:
        # Get colleagues in the same department
        colleagues = df[(df['Department'].str.lower() == dept.lower()) & 
                        (df['Name'].str.lower() != name.lower())]['Name'].tolist()
        colleague_list = ", ".join(colleagues) if colleagues else "No other colleagues"
        msgbox.showinfo("Welcome", f"Welcome {name}!\nYour department colleagues: {colleague_list}")

    else:
        msgbox.showerror("Not Found", f"Sorry, {name} is not an employee in {dept}.")
    
```

```python
# Load employees data
df = pd.read_csv("GIG-logistics(in).csv")

# checking employees data

def c_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    # filtering
    employee = df[(df['Name'].str.lower() == name.lower()) & (df['Department'].str.lower() == dept.lower())]
    if not employee.empty:
        # Get colleagues in the same department
        colleagues = df[(df['Department'].str.lower() == dept.lower()) & 
                        (df['Name'].str.lower() != name.lower())]['Name'].tolist()
        colleague_list = ", ".join(colleagues) if colleagues else "No other colleagues"
        msgbox.showinfo("Welcome", f"Welcome {name}!\nYour department colleagues: {colleague_list}")

    else:
        msgbox.showerror("Not Found", f"Sorry, {name} is not an employee in {dept}.")
    
```

```python
# create main window
root = tk.Tk()
root.title("Employees Verification")
root.geometry("400x200")

# name entry and label
name_label = tk.Label(root, text="Enter your name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# department label and entry

dept_label = tk.Label(root, text="Enter your department:")
dept_label.pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

# submit button
submit_button = tk.Button(root, text="Check Employee", command=c_employee)
submit_button.pack()

# strat event
root.mainloop()
```
