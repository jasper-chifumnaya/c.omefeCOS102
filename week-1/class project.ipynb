```python
#SIMPLE INTEREST
#parameters
p = float(input("Principal?"))
r = float(input("Rate?"))
t = float(input("Time?"))

#total amount to be paid
a = p * (1 + (r / 100) * t)

#interest
i = a - p

print("                           ")
print("Total Amount:", a)
print("Interest to be paid:", i)
```

```python
#COMPOUND INTEREST
#parameters
p = float(input("Principal?"))
r = float(input("Rate?"))
n = float(input("Number of times interest is compounded?"))
t = float(input("Time?"))

#total amount to be paid
a = p * (1 + r / 100) ** ( n * t)

#interest
i = a - p

print("                           ")
print("Total Amount:", a)
print("Interest to be paid:", i)
```

```python
#ANNUITY PLAN

#parameters
pmt = float(input("Periodic Payment?"))
r = float(input("Annual Interest?"))
n = int(input("Number of times interest is compounded?"))
t = float(input("Years?"))

# compute
periodic_rate = r / n           
total_periods = n * t           

#growth
amount_growth = (1 + periodic_rate) ** total_periods

#annuity
a = pmt * ((amount_growth - 1) / periodic_rate)


print("Total Amount (A): ", a)
```
